```python
news_df_columns = ['assetCodes', 'firstCreated', 'wordCount', 'sentenceCount', 
                   'firstMentionSentence', 'sentimentClass',
                   'sentimentNegative', 'sentimentNeutral', 'sentimentPositive',
                   'sentimentWordCount']
news_df = news_df[news_df_columns]
```

```python
def unstack_asset_codes(news_df):
    codes = []
    indexes = []
    for i, values in news_df['assetCodes'].iteritems():
        explode = values.replace('\'', '').replace('}', '').replace('{', '').split(", ")
        codes.extend(explode)
        repeat_index = [int(i)]*len(explode)
        indexes.extend(repeat_index)
    index_df = pd.DataFrame({'news_index': indexes, 'assetCode': codes})
    del codes, indexes 
    return index_df
```

```python
def merge_news_on_index(news_df, index_df):
    news_df['news_index'] = news_df.index.copy()

    # Merge news on unstacked assets
    news_unstack = index_df.merge(news_df, how='left', on='news_index')
    news_unstack.drop(['news_index', 'assetCodes'], axis=1, inplace=True)
    return news_unstack
```

```python
def handle_news_df(news_df):
    news_df = news_df[news_df_columns]
    index_df = unstack_asset_codes(news_df)
    news_df = merge_news_on_index(news_df, index_df)
    news_df = news_df.groupby(['firstCreated', 'assetCode'], as_index=False).mean()
    del index_df
    return news_df
```

```python
def mis_impute(data):
    for i in data.columns:
        if data[i].dtype == "object":
            data[i] = data[i].fillna("other")
        elif (data[i].dtype == "int64" or data[i].dtype == "float64"):
            data[i] = data[i].fillna(data[i].mean())
        else:
            pass
    return data
```

```python
def data_preprocessing_for_training(market_df, news_df):
    # convert to datetime
    market_df['time'] = pd.to_datetime(market_df['time'])
    market_df['time'] = market_df['time'].dt.date
    
    news_df['firstCreated'] = pd.to_datetime(news_df['firstCreated'])
    news_df['firstCreated'] = news_df['firstCreated'].dt.date
    news_df = handle_news_df(news_df)
    full_df = pd.merge(market_df, news_df, how='left', left_on=['time', 'assetCode'], right_on=['firstCreated', 'assetCode'])
    label = full_df.returnsOpenNextMktres10 >= 0
    full_df = full_df.drop(['assetCode', 'time', 'assetName', 'firstCreated', 'returnsOpenNextMktres10', 'binary_returnsNextMktres10'], axis=1)
    full_df = mis_impute(full_df)
    return full_df, label
```

```python
X, label_train = data_preprocessing_for_training(market_df, news_df)
```

```python
X.describe()
```

```python
import numpy as np
X = X.values
# Scaling the X values
mins = np.min(X, axis=0)
maxs = np.max(X, axis=0)
rng = maxs - mins
X = 1 - ((maxs - X) / rng)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, label_train, test_size=0.2, random_state=29)
```

```python
import lightgbm as lgb
train_data = lgb.Dataset(X_train, label=y_train.astype(int))

x = [0.19000424246380565, 2452, 212, 328, 202]

params = {
       'task': 'train',
       'boosting_type': 'gbdt',
       'objective': 'binary',
       'learning_rate': x[0],
       'num_leaves': x[1],
       'min_data_in_leaf': x[2],
       'num_iteration': 239,
       'max_bin': x[4],
       'verbose': 1
   }


gbm = lgb.train(params,
       train_data,
       num_boost_round=100,
       )
```

```python
from sklearn.metrics import accuracy_score
res = gbm.predict(X_test) 

# calculating confidence value
confidence = res
confidence = (confidence-confidence.min())/(confidence.max()-confidence.min())
confidence = confidence * 2 - 1
```

```python
# Accuracy of the model predicting whether the 'returnsOpenNextMktres10' column is greater than zero 

res2 = []
for tmp in res:
    if(tmp<0.5): 
        res2.append(0)
    else: 
        res2.append(1)

print(accuracy_score(res2, y_test))
```
